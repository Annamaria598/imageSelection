## Valori da indagare ulteriormente
*WINDOW_SIZE*: Tutti i migliori risultati hanno una window_size di 30 pixel. Continuiamo a mantenere questo valore ed indaghiamo valori intorno a questo. Es: 20, 25, 30, 35, 40.

*Neighbors*: Nei migliori 25 risultati appaiono prevalenti i valori 5, 10, 25 ma senza una differenza eccessiva. Proviamo ad ampliare. Es: 3, 5, 7, 10, 15, 20, 25, 50

*Features*: Proviamo a mantenere i primi 10 migliori insiemi di features e creiamo nuovi insiemi mescolando valori calcolati in spazi colore differenti. L'entropia è presente nei modelli con i risultati migliori. È il caso di tenerla.

```python
from itertools import combinations
```

```python
# rimuovo tutte le combinazioni in cui tutte le features
# sono calcolate sullo stesso spazio colore
def stesso_spazio(features):
    
    lab = ["L", "a", "b"]
    rgb = ["R", "G", "B"]
    hsv = ["h", "s", "v"]
    
    in_lab = 0
    in_rgb = 0
    in_hsv = 0
    
    for f in features:
        element = f.split("_")[0]
        if element in lab:
            in_lab = 1
        elif element in rgb:
            in_rgb = 1
        else:
            in_hsv = 1
            
        if in_lab + in_rgb + in_hsv > 1:
            return False
            
    return True
```

```python
features_to_mix = [
    ['R_mf', 'G_mf', 'B_mf'],
    ['R_variance', 'G_variance', 'B_variance'],
    ['R_edge', 'G_edge', 'B_edge'],
    ['L_mf', 'a_mf', 'b_mf'],
    ['L_variance', 'a_variance', 'b_variance'],
    ['L_edge', 'a_edge', 'b_edge'],
    ['h_mf', 's_mf', 'v_mf'],
    ['h_variance', 's_variance', 'v_variance'],
    ['h_edge', 's_edge', 'v_edge']
]
```

```python
comb2 = combinations(features_to_mix, 2)
```

```python
comb3 = combinations(features_to_mix, 3)
```

```python
from itertools import chain
```

```python
comb2_lists = []
for c in comb2:
    comb2_lists.append(list(chain.from_iterable(c)))
```

```python
comb2_lists[:2]
```

```python
comb3_lists = []
for c in comb3:
    comb3_lists.append(list(chain.from_iterable(c)))
```

```python
comb3_lists[:2]
```

```python
comb = comb2_lists + comb3_lists

# rimuovo set formati da features da un unico spazio
comb = [c for c in comb if not stesso_spazio(c)]
```
